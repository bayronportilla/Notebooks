# Graficador basico


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
T=2.0*np.pi
time = np.linspace(0,2.0*T,1000)
```

```python
plt.figure
plt.plot(time,np.sin(0.3*time))
plt.plot(time,0.5*np.sin(10.0*time))
plt.xlabel("tiempo")
plt.ylabel("propiedad")
plt.show()
```
